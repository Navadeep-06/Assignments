```python
import pandas as pd

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder #with encoder

# Initialize LabelEncoder
le=LabelEncoder()

# Apply Label Encoding to the 'Gender' column by creating a new gender_encoded column
df["Gender"]=le.fit_transform(df["Gender"])

# Display the updated DataFrame
print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False) # sparse_output=False makes it return a dense array directly

# Fit and transform the 'City' column
# ohe expects a 2D array-like input, df[['City']] is a DataFrame (2D)
encoded_cities_array = ohe.fit_transform(df[['City']])

# Get feature names for the new columns
feature_names = ohe.get_feature_names_out(['City'])

# Create a DataFrame from the encoded data
encoded_df = pd.DataFrame(encoded_cities_array, columns=feature_names, index=df.index)

# Concatenate the original DataFrame (without 'City') with the new encoded DataFrame
df = pd.concat([df.drop('City', axis=1), encoded_df], axis=1)

df
```

```python
df['Size']=le.fit_transform(df['Size'])
df
```
